```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
m=100
T_m=70+np.random.random(m)
t=np.arange(m)+1
```

```python
T=[70.00]
E_st=[0.05]
E_mea=0.2
for i in range(m):
    K_G=E_st[i]/(E_st[i]+E_mea)
    T.append(T[i]+K_G*(T_m[i]-T[i]))
    E_st.append((1-K_G)*E_st[i])
```

```python
plt.plot(np.ones(m)*T_m.mean())
plt.plot(T)
plt.scatter(t,T_m)
plt.show()
```
